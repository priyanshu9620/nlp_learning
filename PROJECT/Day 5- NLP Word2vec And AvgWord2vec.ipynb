```python
# importing the Dataset

import pandas as pd
import numpy as np

messages = pd.read_csv('smsspamcollection/SMSSpamCollection', sep='\t',
                           names=["label", "message"])
# we have seperated our data using \t tab space and given the columns some name 
```

```python
# now road map 
# 1. text pre-> tokenization, stopwords,stemming, lemmatization
# 2. text->vectors -> bow,TFIDF , word2vec,avgword2vec
```

```python
#Data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#we are applying stemming but u can also do lemetization 
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
#     this join means that when ever u have space join jaise hi comma aye then next line 
    corpus.append(review)
```

```python
corpus
```

# Using BOW 

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,binary=True,ngram_range=(2,2))
X = cv.fit_transform(corpus).toarray()
```

```python
y=pd.get_dummies(messages['label'])
#first comment out this "iloc" line and see y then use iloc and values to convert that to array
# see iloc just seperates based on index which u give so here we have given 
# index 1 so it is seperating based on index 1 which is spam and showing weather it
# is true or false and .values is converting the output into an array
y=y.iloc[:,1].values
print(y)
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
```

```python
#prediction
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```

# now we did bow and got this acc lets now try same thing using TFIDF

```python
# Creating the TFIDF model
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=2500)
X = tv.fit_transform(corpus).toarray()
```

```python
#prediction
y_pred=spam_detect_model.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
#in nlp2 we used the google word2vec model and here we are training our model
# from scratch. firstly i used lemmatization instead of stemming in this case 
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
# difference between lemma and stemming is in prev case my corpus has meaning
# less words and in this case it will have all meaningful words
corpus[0]
```

```python
# SIMPLE TOKENIZATION 
# this simple preprocess function just convert them to lower cases: u can also do sentence.to_lower()
# and also ignore words which are either two long or short 
words=[]
for sentence in corpus:
    sent_token=sent_tokenize(sentence)
    for sentence in sent_token:
        words.append(simple_preprocess(sentence))
```

```python
words
```

```python
# now data preprocessing is done creating our word2vec from strach
import gensim
```

```python
model=gensim.models.Word2Vec(words,window=5,min_count=2)
#min count just ignore all the words whose freq is lower than this 
```

```python
#this is my complete vocab
model.wv.index_to_key
```

```python
len(model.wv['kid']) #so just showing len of the vector 
#this word to vec each word has dimension of 100 now we will go for avg word to vec
```

```python
def avg_word2vec(doc):
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
        
# simple funciton just finding avg if that word was in vocab 
    
```

```python
# now funciton is created and we are gonna apply that to our words
words
```

```python
#apply for the entire sentences
X=[]
for i in tqdm(range(len(words))):
    X.append(avg_word2vec(words[i]))

    
```
